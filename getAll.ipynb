```python
# 获取每一个房子的大概信息
from urllib import request
from bs4 import BeautifulSoup
import pandas
import requests
import traceback

prefix = 'https://www.haozu.com/sh/'
local = ['pudongxinqu','huangpu','xuhui','changning','jingan','putuo','zhabei'
          ,'hongkou','yangpu','minhang','baoshan','jiading','songjiang'
          ,'qingpu','fengxian','shanghaizhoubian']
# 保存所有的网络地址
houseLink1 = [prefix + 'house-' + item + '/a1/' for item in local] # 普通办公、
houseLink2 = [prefix + 'house-' + item + '/a6/' for item in local] # 创意园区

# 保存所有的excel的名字
excelName1 = ['content-a1-' + item + '.xlsx' for item in local] # 普通办公
excelName2 = ['content-a6-' + item + '.xlsx' for item in local] # 创意园区
# 获取设置过ua 和 ip 的request
def getRequest():
    # proxy = {'https':'119.101.117.215:9999'}
    proxy = {'https':'119.101.112.71:9999'}
    proxy_support = request.ProxyHandler(proxy)
    opener = request.build_opener(proxy_support)
    # opener.addheaders = [('User-Agent','Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36')]
    opener.addheaders = [('User-Agent','Mozilla/4.0 (compatible; MSIE 6.0; Windows NT 5.1)')]
    request.install_opener(opener)
    return request

# 1 根据房子的类型，获取此类型下的每一页地址
def getFirstUrl(prefix):
    allUrl = [prefix]
    for item in range(2, 10000) : 
        temp = str(prefix) + 'o' + str(item) + '/'
        allUrl.append(temp)
    return allUrl

# 信息
# 2 传进列表prefixList: 根据这个类型下的所有页面地址,获取每个页面的所有的房子的信息;
#   传进一个字符串excelName: excel的名字
def getSecondUrl(prefixList,excelName):
    all = [] # 保存每一个房源的信息，嵌套着字典
    hrefs = [] # 保存每一个房源的地址
#     req = getRequest()
    try:
        page = 0
        for url in prefixList:
#             response = req.urlopen(url)
#             html = response.read().decode("utf-8")
            html = requests.get(url).text
            soup = BeautifulSoup(html,'html.parser')
            content = soup.select('.list-content')
            if content:
                for item in content:
                    all.append({'content':item})
                page += 1
                print(len(all),end = '  ') # 输出爬了多少条数据了
            else:
                print('\n正常结束', '；类型：', excelName, '；页数：', page, '数据量：',len(all))
                break
    except Exception as e:
        print("\n获取信息过程出现错误")
        traceback.print_exc()
    finally:
        # 将所有的信息保存进excel
        all = pandas.DataFrame(all)
        all.to_excel('./content/'+ excelName)
        print('信息保存进excel成功！') 
def mymain():
    i = 0
    for url, name in zip(houseLink1, excelName1) :
        i += 1
        print('第 ',i,' 次开始遍历') 
        allPage = getFirstUrl(url)
        allUrl = getSecondUrl(allPage,name) 
mymain()
```
